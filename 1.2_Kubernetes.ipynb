# Principios de Kubernetes

Kubernetes es una plataforma de orquestación de contenedores de código abierto que automatiza la implementación, el escalado y la gestión de aplicaciones en contenedores. Kubernetes proporciona a los desarrolladores un marco para ejecutar fácilmente sistemas distribuidos. Kubernetes también ofrece a los desarrolladores opciones y flexibilidad a la hora de crear plataformas.

En esta lectura, aprenderá más sobre los principios y la filosofía de diseño detrás de Kubernetes y aprenderá cómo la configuración declarativa permite el modelo de reconciliación de estados deseado. Además, aprenderá sobre los componentes clave del plano de control.

## Principios de Kubernetes

Kubernetes, una aplicación nativa de la nube, sigue principios para garantizar que la aplicación en contenedores se ejecute correctamente. Estos principios tienen en cuenta el tiempo de construcción y el tiempo de ejecución. Un contenedor es autónomo y depende únicamente del kernel de Linux. Una vez creado el contenedor, se pueden agregar bibliotecas adicionales. Además, las aplicaciones en contenedores no están destinadas a cambiar a diferentes entornos una vez creadas.

En términos de tiempo de ejecución, cada contenedor necesita implementar API para ayudar a la plataforma a administrar la aplicación de la manera más eficiente posible. Todas las API deben ser públicas, ya que no debe haber API ocultas o privadas. Además, las API deben ser declarativas, lo que significa que el programador debe poder comunicar el resultado final deseado, lo que permitirá a Kubernetes encontrar e implementar una solución. Hay soporte disponible para los desarrolladores, si es necesario, para ejecutar aplicaciones en Kubernetes. Las cargas de trabajo son portátiles y el plano de control puede transferir una carga de trabajo a otro nodo sin interrumpir el programa general.

## Configuración declarativa

La configuración declarativa es un enfoque que se usa comúnmente en Kubernetes para lograr el estado deseado de una aplicación. En este enfoque, los desarrolladores especifican el estado deseado, pero no definen explícitamente cómo lograrlo o alcanzarlo. El enfoque se centra más en cuál debería ser el estado deseado. El sistema determinará la forma más eficiente y confiable de lograr el estado deseado. Estos activos de configuración se almacenan en un sistema de control de revisiones y rastrean los cambios a lo largo del tiempo.

Para utilizar la configuración declarativa en Kubernetes, cree un manifiesto que describa el estado deseado de una aplicación. Luego, el plano de control determinará cómo dirigir los nodos en el clúster para lograr el estado deseado.

### el plano de control

El plano de control de Kubernetes es responsable de tomar decisiones sobre todo el clúster y el estado deseado y de garantizar que los componentes del clúster funcionen juntos. Los componentes del plano de control incluyen: 

- etcd

- API Server

- Scheduler

- Controller manager

- Cloud Controller manager

**etcd** se utiliza como almacén de respaldo de Kubernetes para todos los datos del clúster como una base de datos distribuida. Este almacén clave-valor tiene alta disponibilidad y está diseñado para ejecutarse en múltiples nodos.

El **API Server** de Kubernetes actúa como interfaz para los desarrolladores y otros componentes que interactúan con el clúster. Es responsable de garantizar que las solicitudes a la API estén debidamente autenticadas y autorizadas.

El **programador** (Scheluder) es un componente del plano de control donde se asignan pods para que se ejecuten en nodos particulares del clúster.

El **administrador de control** aloja varios controladores de Kubernetes. Cada controlador monitorea continuamente el estado actual del clúster y trabaja para lograr el estado deseado.

El **administrador del controlador de la nube** es un componente del plano de control que incorpora una lógica de control específica de la nube. Actúa como interfaz entre Kubernetes y un proveedor de nube específico, gestionando los recursos de la nube.

### Conclusiones clave

Kubernetes es una plataforma portátil y extensible para ayudar a los desarrolladores con aplicaciones en contenedores. Los principios básicos y los componentes clave de Kubernetes ayudan a los desarrolladores a iniciar, detener, almacenar, crear y administrar contenedores.

# Pods

En Kubernetes, un contenedor es un paquete ejecutable, liviano e independiente que incluye todo lo necesario para ejecutar una pieza de software, incluido el código, un tiempo de ejecución, bibliotecas, variables de entorno y herramientas del sistema. Los contenedores están aislados entre sí y agrupan su propio software, bibliotecas y archivos de configuración, pero comparten el núcleo del sistema operativo con otros contenedores. Están diseñados para ser fácilmente portátiles en diferentes entornos, lo que los hace ideales para una implementación consistente en diferentes plataformas.

En el contexto de Kubernetes, los contenedores son las unidades de implementación más pequeñas que se programan y administran. Están encapsulados dentro de Pods , que son las unidades de implementación fundamentales en un clúster de Kubernetes. Un Pod puede contener uno o más contenedores que deben ejecutarse juntos en el mismo host y compartir la misma red y recursos de almacenamiento, lo que les permite comunicarse entre sí mediante localhost.

Los pods sirven como una capa de abstracción, lo que permite a Kubernetes programar y organizar contenedores de manera efectiva. Cuando una implementación requiere que varios contenedores trabajen juntos en el mismo nodo, se crea un Pod para garantizar que estén ubicados juntos y puedan comunicarse de manera eficiente. Esto simplifica la implementación y administración de aplicaciones en contenedores, lo que facilita su escalado, monitoreo y actualización según sea necesario.

Tenga en cuenta también que los Pods en Kubernetes se consideran efímeros; se pueden crear, finalizar y reemplazar dinámicamente según el estado deseado y la disponibilidad de recursos en el clúster. Como resultado, Kubernetes garantiza que siempre se esté ejecutando la cantidad deseada de Pods, lo que permite una alta disponibilidad y tolerancia a fallas para las aplicaciones en contenedores.

Los pods sirven juntos como un host lógico que encapsula uno o más contenedores estrechamente acoplados dentro de una red compartida y un contexto de almacenamiento. Esto proporciona una manera de agrupar contenedores que necesitan trabajar en estrecha colaboración, permitiéndoles compartir los mismos recursos e interactuar entre sí como si se estuvieran ejecutando en la misma máquina física o virtual.

Al diseñar un Pod, considere aspectos como las solicitudes y los límites de recursos, el manejo de apagados ordenados, el registro y la supervisión, y las imágenes de contenedor adecuadas.

## Pods como host lógico

Un Pod puede ejecutar uno o más contenedores estrechamente relacionados que comparten la misma red y contexto de almacenamiento. Este contexto compartido es muy parecido al que encontraría en una máquina física o virtual, de ahí el término "host lógico". 

Los puntos clave que hay que entender acerca de un Pod como host lógico son:

- Contenedores estrechamente acoplados: cuando varios contenedores dentro de un Pod se consideran estrechamente acoplados, significa que tienen una fuerte interdependencia y necesitan comunicarse entre sí a través del host local. Esto les permite intercambiar datos e información de manera eficiente sin la necesidad de configuraciones de red complejas.

- Espacio de nombres de red compartido: los contenedores dentro del mismo Pod comparten el mismo espacio de nombres de red. Esto implica que tienen la misma dirección IP y espacio de puerto, lo que les facilita la comunicación mediante mecanismos estándar de comunicación entre procesos.

- Contexto de almacenamiento compartido: los pods también comparten el mismo contexto de almacenamiento, lo que significa que pueden acceder a los mismos volúmenes o recursos de almacenamiento. Esto facilita el intercambio de datos entre los contenedores dentro del Pod, mejorando aún más su colaboración.

- Co-ubicación y co-programación: Kubernetes garantiza que todos los contenedores dentro de un Pod estén programados y co-ubicados en el mismo nodo. Esta programación conjunta garantiza que los contenedores puedan comunicarse eficientemente entre sí dentro de la misma red y contexto de almacenamiento.

- Naturaleza efímera: al igual que los contenedores individuales, los Pods se consideran efímeros y pueden crearse, finalizarse o reemplazarse fácilmente según los requisitos de escala o las limitaciones de recursos. Sin embargo, todos los contenedores dentro del Pod se tratan como una sola unidad en términos de programación y gestión del ciclo de vida.

## Pods en acción

Supongamos que es un desarrollador de software a cargo de una aplicación web que incluye un servidor web principal y un componente auxiliar para el procesamiento de registros. El servidor web interactúa con el procesador de registros para manejar, analizar y almacenar datos de registros en tiempo real. Estos dos componentes deben estar estrechamente integrados y comunicarse entre sí de manera eficiente.

En este escenario, usaría un Pod de Kubernetes para encapsular tanto el servidor web como los contenedores del procesador de registros. Dado que ambos contenedores existen dentro del mismo Pod, comparten el mismo espacio de nombres de red (pueden comunicarse a través de localhost) y pueden compartir los mismos volúmenes de almacenamiento. Esto permite que el servidor web genere registros y que el procesador de registros acceda y procese estos registros de manera eficiente.

El Pod garantiza que tanto el servidor web como el procesador de registros estén programados en el mismo nodo (ubicación conjunta) y se administren como una sola entidad. Si es necesario reprogramar el Pod o si falla, ambos contenedores se tratarían juntos, manteniendo su relación de acoplamiento. El Pod abstrae los detalles de la máquina host y la infraestructura subyacente, lo que le permite concentrarse en administrar su aplicación.

Esta configuración, en la que se agrupan varios contenedores relacionados en un Pod, se conoce como Pod de contenedores múltiples. Explorará los pods de contenedores únicos y múltiples con más detalle a continuación; Por ahora, sólo sepa que varios contenedores son una forma ideal de gestionar e implementar componentes de aplicaciones estrechamente acoplados. 

## Ventajas de los Pods

En el ejemplo anterior, puede ver que los Pods ofrecen una serie de ventajas al facilitar la ubicación conjunta de contenedores, permitir el intercambio de datos y simplificar la comunicación entre contenedores:

- Facilitar la ubicación conjunta: los pods permiten la ubicación conjunta de varios contenedores en la misma máquina host. Esto es particularmente útil para componentes estrechamente relacionados que necesitan trabajar juntos, como una aplicación y sus componentes auxiliares (como contenedores secundarios que manejan el registro o la supervisión). Al ejecutar estos componentes en el mismo Pod, se pueden programar en la misma máquina y administrar como una sola entidad.

- Habilitar el intercambio de datos: los contenedores dentro de un Pod comparten el mismo espacio de nombres de red, lo que significa que comparten una dirección IP y un espacio de puerto. Pueden comunicarse entre sí mediante localhost y también pueden compartir datos a través de volúmenes compartidos. Los volúmenes compartidos en un Pod permiten que los datos se intercambien fácilmente entre contenedores y también permiten que los datos persistan más allá de la vida útil de un único contenedor, lo que puede resultar útil para aplicaciones que requieren almacenamiento de datos persistente.

- Simplificación de la comunicación entre contenedores: el espacio de nombres de red compartido también simplifica la comunicación entre contenedores. Debido a que todos los contenedores en un Pod comparten una pila de red, pueden comunicarse entre sí en localhost, sin la necesidad de comunicación entre procesos (IPC) o sistemas de archivos compartidos. Esto simplifica el desarrollo de sistemas distribuidos, donde los componentes a menudo necesitan comunicarse entre sí.

## Contenedor único versus contenedores múltiples

La diferencia entre Pods de un solo contenedor y de múltiples contenedores radica en la cantidad de contenedores que albergan.

Como sugiere el nombre, los Pods de un solo contenedor contienen solo un contenedor. Este contenedor normalmente representa la aplicación o servicio principal que el Pod debe ejecutar. Los Pods de un solo contenedor son sencillos y se usan comúnmente cuando se tiene una aplicación simple que no requiere contenedores adicionales ni componentes auxiliares estrechamente relacionados. También son adecuados para ejecutar aplicaciones independientes que no necesitan comunicarse con otros contenedores dentro del mismo Pod. 

Los pods de contenedores múltiples , por otro lado, contienen múltiples contenedores que están ubicados en el mismo lugar y comparten los mismos recursos y espacio de nombres de red. Estos contenedores están destinados a trabajar juntos y complementar las funcionalidades de cada uno. Los Pods de contenedores múltiples son apropiados en varios escenarios:  	

- Patrón sidecar: el patrón sidecar es un caso de uso común para Pods de contenedores múltiples. En este patrón, el contenedor principal representa la aplicación principal, mientras que los contenedores adicionales proporcionan funciones de soporte como registro, monitoreo o autenticación. Los contenedores sidecar mejoran y amplían las capacidades de la aplicación principal sin modificar su código.

- Patrón de proxy: los pods de contenedores múltiples pueden usar un contenedor proxy que actúa como intermediario entre el contenedor principal de la aplicación y el mundo externo. El contenedor proxy maneja tareas como equilibrio de carga, almacenamiento en caché o terminación SSL, descargando estas responsabilidades del contenedor principal de la aplicación.

- Patrón de adaptador: los pods de contenedores múltiples pueden emplear un contenedor de adaptador que realiza conversiones de formatos de datos o traducciones de protocolos. Esto permite que el contenedor principal se centre únicamente en su funcionalidad principal sin preocuparse por las complejidades de los formatos de intercambio de datos.

- Datos compartidos y dependencias: los contenedores dentro de un Pod de múltiples contenedores pueden compartir volúmenes y comunicarse a través del host local, lo que los hace adecuados para aplicaciones que requieren compartir datos o tienen componentes interdependientes.

Utilice un Pod de contenedor único cuando tenga una aplicación simple que no requiera contenedores adicionales, o cuando desee aislar diferentes aplicaciones o servicios para facilitar su administración y escalado. 

Utilice Pods de contenedores múltiples cuando tenga componentes estrechamente relacionados que necesiten trabajar juntos, como los que siguen el patrón sidecar. Esto es útil para tareas como registrar, monitorear o mejorar las capacidades de la aplicación principal sin modificar su código. Los pods de contenedores múltiples también son apropiados para escenarios en los que varios contenedores necesitan compartir datos o dependencias de manera eficiente.

## Términos clave

A continuación se incluyen algunos términos clave con los que debe familiarizarse mientras trabaja con Kubernetes.

- Ciclo de vida del pod: los pods tienen fases de ciclo de vida específicas, que van desde "Pendiente" cuando se están programando, hasta "En ejecución" cuando todos los contenedores están en funcionamiento, "Exitoso" cuando todos los contenedores finalizan exitosamente y "Error" si algún contenedor dentro del El pod no se ejecuta. Los pods también pueden estar en estado "ContainerCreating" si se están creando uno o más contenedores.

- Plantillas de pod: las plantillas de pod definen la especificación para crear nuevos pods. Se utilizan en controladores de nivel superior como ReplicaSets, Deployments y StatefulSets para garantizar el estado deseado de los Pods.

- Afinidad y antiafinidad de pods: las reglas de afinidad y antiafinidad de pods definen las preferencias y restricciones de programación de los pods. Le permiten influir en la coubicación o separación de Pods en función de etiquetas y otros atributos.

- Escalado automático de pods: Kubernetes proporciona la funcionalidad de escalador automático de pods horizontal (HPA) que escala automáticamente la cantidad de réplicas (Pods) en función del uso de recursos o métricas personalizadas.

- Políticas de seguridad de Pod: las políticas de seguridad de Pod se utilizan para controlar los aspectos relacionados con la seguridad de los Pods, como su acceso a ciertos recursos del host, el uso de contenedores privilegiados y más.

- Contenedores de inicio: los contenedores de inicio son contenedores adicionales que se ejecutan y completan antes de que comiencen los contenedores de la aplicación principal. Son útiles para realizar tareas de inicialización, como la configuración del esquema de la base de datos o la precarga de datos.

- Desalojamiento e interrupción de pods: los pods pueden ser desalojados de los nodos debido a limitaciones de recursos o fallas en los nodos. Comprender el comportamiento de desalojo de Pod es importante para gestionar la confiabilidad de las aplicaciones.

- Sondas de estado del pod: Kubernetes admite diferentes tipos de sondeos de estado (sondas de actividad, preparación y de inicio) para verificar el estado de los contenedores dentro de un Pod. Estas pruebas ayudan a Kubernetes a decidir si un Pod se considera en buen estado y listo para recibir tráfico.

- Contaminaciones y tolerancias: las contaminaciones se aplican a los nodos para repeler los Pods, mientras que las tolerancias se establecen en los Pods para permitir que se programen en los nodos contaminados.

- DNS de pod: a los pods se les asigna un nombre de host y una dirección IP únicos. Pueden comunicarse entre sí utilizando su nombre de host o nombres de servicio. Kubernetes proporciona resolución DNS interna para facilitar la comunicación entre Pods.

- Anotaciones y etiquetas de pods: se pueden adjuntar anotaciones y etiquetas a los pods para proporcionar metadatos o facilitar la selección de pods para diversos fines, como monitoreo, registro o enrutamiento.

## Pods y Python

Para administrar pods de Kubernetes usando Python, puede usar la biblioteca de Kubernetes. A continuación se muestra un código de ejemplo sobre cómo crear, leer, actualizar y eliminar un Pod usando Python.

```python
from kubernetes import client, config

# Load the Kubernetes configuration from the default location
config.load_kube_config()

# Alternatively, you can load configuration from a specific file
# config.load_kube_config(config_file="path/to/config")

# Initialize the Kubernetes client
v1 = client.CoreV1Api()

# Define the Pod details
pod_name = "example-pod"
container_name = "example-container"
image_name = "nginx:latest"
port = 80

# Create a Pod
def create_pod(namespace, name, container_name, image, port):
	container = client.V1Container(
    	name=container_name,
    	image=image,
    	ports=[client.V1ContainerPort(container_port=port)],
	)

	pod_spec = client.V1PodSpec(containers=[container])
	pod_template = client.V1PodTemplateSpec(
    	metadata=client.V1ObjectMeta(labels={"app": name}), spec=pod_spec
	)

	pod = client.V1Pod(
    	api_version="v1",
    	kind="Pod",
    	metadata=client.V1ObjectMeta(name=name),
    	spec=pod_spec,
	)
	
	try:
		response = v1.create_namespaced_pod(namespace, pod)
    	print("Pod created successfully.")
		return response
	except Exception as e:
    	print("Error creating Pod:", e)


# Read a Pod
def get_pod(namespace, name):
	try:
    	response = v1.read_namespaced_pod(name, namespace)
    	print("Pod details:", response)
	except Exception as e:
    	print("Error getting Pod:", e)


# Update a Pod (e.g., change the container image)
def update_pod(namespace, name, image):
	try:
    	response = v1.read_namespaced_pod(name, namespace)
    	response.spec.containers[0].image = image

    	updated_pod = v1.replace_namespaced_pod(name, namespace, response)
    	print("Pod updated successfully.")
    	return updated_pod
	except Exception as e:
    	print("Error updating Pod:", e)


# Delete a Pod
def delete_pod(namespace, name):
	try:
    	response = v1.delete_namespaced_pod(name, namespace)
    	print("Pod deleted successfully.")
	except Exception as e:
    	print("Error deleting Pod:", e)


if __name__ == "__main__":
	namespace = "default"

	# Create a Pod
	create_pod(namespace, pod_name, container_name, image_name, port)

	# Read a Pod
	get_pod(namespace, pod_name)

	# Update a Pod
	new_image_name = "nginx:1.19"
	update_pod(namespace, pod_name, new_image_name)

	# Read the updated Pod
	get_pod(namespace, pod_name)

	# Delete the Pod
	delete_pod(namespace, pod_name)
```

Los pods sirven como una capa de abstracción, lo que permite a Kubernetes programar y organizar contenedores de manera efectiva.

Utilice un Pod de contenedor único cuando tenga una aplicación simple que no requiera contenedores adicionales, o cuando desee aislar diferentes aplicaciones o servicios para facilitar su administración y escalado. 

Utilice Pods de contenedores múltiples cuando tenga componentes estrechamente relacionados que necesiten trabajar juntos, como los que siguen el patrón sidecar.
